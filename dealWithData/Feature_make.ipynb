```python
import pandas as pd
header_list = ['UID', 'Date', 'Time', 'In', 'Out', 'Number']
datList = ['201702.csv','201703.csv','201704.csv','201705.csv','201706.csv','201707.csv','201708.csv','201709.csv','201710.csv','201711.csv','201712.csv','201801.csv','201802.csv','201803.csv','201804.csv']
df_final = pd.read_csv('non-normal.csv',index_col = 0)
for k in datList:
    path = 'cleanData/' + k
    df = pd.read_csv(path,index_col = 'UID',names=header_list)
    df.drop(df.index[0],inplace=True)
    dateList = []
    timeList = []
    dateList = sorted(list(set(df['Date'].values.tolist())))
    timeList = sorted(list(set(df['Time'].values.tolist())))
    for i in dateList:
        for j in timeList:
            lst = []
            df_head = []
            lst_temp = []
            date = []
            time = []
            df_temp = df[(df.Time == j) & (df.Date == i) & (df.Out == '淡水')]
            df_head = df_temp['In'].values.tolist()
            if df_head:
                df_head.append('Time')
                df_head.append('Date')
                df_head.append('Week')
                df_head.append('Total')
                lst_temp = df_temp['Number'].values.tolist()
                print(lst_temp)
                date = df_temp['Date'].values[0]
                print('Date:' + date)
                time = df_temp['Time'].values[0]
                print('Time:' + time)
                week = datetime.datetime(int(date.split('-')[0]),int(date.split('-')[1]),int(date.split('-')[2])).strftime("%w");
                print('Week:' + week)
                lst_temp.append(time)
                lst_temp.append(date)
                lst_temp.append(week)
                num_total = df_temp['Number'].values.sum()
                print(num_total)
                lst_temp.append(num_total)
                lst.append(lst_temp)
                df_temp = pd.DataFrame(lst,columns = df_head)
                df_final = df_final.append(df_temp, ignore_index=True)
    print('Finished a month')
print('All finished , start export the data.')
df_final.to_csv('non-normal_all.csv', encoding='utf-8')
print('Export data finished.')
```
